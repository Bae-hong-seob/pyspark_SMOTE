```python
import pyspark
from pyspark.sql import *
from pyspark.sql.types import *
import pyspark.sql.functions as F
from pyspark import SparkContext, SparkConf


sc=pyspark.SparkContext()
# spark = SparkSession.builder.getOrCreate()

spark = SparkSession\
    .builder\
    .config("spark.driver.memory", "16g")\
    .config("spark.executor.memory", "4g")\
    .getOrCreate()


import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
from google.colab import drive
drive.mount('/content/drive')

train_data = spark.read.csv("/content/drive/MyDrive/빅데이터 처리 및 응용/팀플/train_data.csv", header=True, inferSchema=True)
test_data = spark.read.csv("/content/drive/MyDrive/빅데이터 처리 및 응용/팀플/test_data.csv", header=True, inferSchema=True)
```

# RandomUnderSampling 알고리즘 구현

```python
# 주 클래스와 소수 클래스의 비율 계산
origin_major_df = train_data.filter(train_data.HeartDisease_Index == 0.0)
origin_minor_df = train_data.filter(train_data.HeartDisease_Index == 1.0)

ratio = int(origin_major_df.count()/origin_minor_df.count())
print("ratio: {}".format(ratio))
a = range(ratio)

# 다수 클래스 샘플링해서 제거
major_df = train_data.filter(train_data.HeartDisease_Index == 0.0)
minor_df = train_data.filter(train_data.HeartDisease_Index == 1.0)

sampled_majority_df = major_df.sample(False, 1/ratio)
undersample_train_data = sampled_majority_df.unionAll(minor_df)
undersample_train_data.show()

undersampled_major_df = undersample_train_data.filter(undersample_train_data.HeartDisease_Index == 0.0)
undersampled_minor_df = undersample_train_data.filter(undersample_train_data.HeartDisease_Index == 1.0)

# 데이터 비율 및 수
print("샘플링 전 데이터 개수: {}".format(train_data.count()))
print("샘플링 후 0인 경우 : {}".format(major_df.count()))
print("샘플링 후 1인 경우 : {}".format(minor_df.count()))
print()
print()
print("언더 샘플링 후 데이터 전체수: {}".format(undersample_train_data.count()))
print("언더 샘플링 데이터 label 0인 경우 : {}".format(undersampled_major_df.count()))
print("언더 샘플링 데이터 label 1인 경우 : {}".format(undersampled_minor_df.count()))

```
